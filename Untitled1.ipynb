```python
def extract_paragraphs(text_final):
    paragraph_regex = r'\n\s*\n'
    paragraphs = re.split(paragraph_regex, text_final)
    return paragraphs
print(paragraphs)
```

```python
# ADP + NOUN
# NOUN + ADJ
# VERB + NOUN
# NOUN + CCONJ + NOUN
# DET + NOUN


# Если вводить текст строкой, получается находить нужнеы сочетания

import spacy
from spacy.matcher import Matcher
nlp = spacy.load("ru_core_news_sm")

text = "Воспитание детей и технологии. У родителей много вопросов о цифровом будущем. Какие технологии плохие, а какие хорошие? Что полезно для ребёнка, а что идёт ему во вред? Непростые вопросы, на которые наше общество пытается найти ответ. Я всегда восхищалась семьями, открытыми новым технологиям, возможно, потому что я воспитывалась матерью, у которой не было вообще никакой техники. Когда я стала социальным психологом в 1980 году, я спрашивала семьи о новых проблемах того времени. В то время вопросы были такими: «Куда поставить компьютер?» «Стоит ли ставить телевизор в детских спальнях?» Родителей беспокоили насилие и реклама."
doc = nlp(text)
matcher_1 = Matcher(nlp.vocab)
pattern_1 = [{"POS": "ADP"}, {"POS": "NOUN"}]
matcher_1.add("AdpNoun", [pattern_1])
matches = matcher_1(doc)
for match_id, start, end in matches:
    span_1 = doc[start:end]
    print(span_1)
```

```python
matcher_2 = Matcher(nlp.vocab)
pattern_2 = [{"POS": "NOUN"}, {"POS": "ADJ"}]
matcher_2.add("NounAdj", [pattern_2])
matches = matcher_2(doc)
for match_id, start, end in matches:
    span_2 = doc[start:end]
    print(span_2)
```
